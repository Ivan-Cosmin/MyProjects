```python
import mediapipe as mp
import cv2
import numpy as np
```

```python
def normalize_landmark_list(landmark_list):
    base_x = landmark_list[0].x
    base_y = landmark_list[0].y
    base_z = landmark_list[0].z

    for landmark in landmark_list:
        landmark.x -= base_x
        landmark.y -= base_y
        landmark.z -= base_z

    landmark_list[0].x = 0
    landmark_list[0].y = 0
    landmark_list[0].z = 0

    flattened_landmark_list = []
    for landmark in landmark_list:
        flattened_landmark_list.append(landmark.x)
        flattened_landmark_list.append(landmark.y)
        flattened_landmark_list.append(landmark.z)
    
    #print(landmark_list)

    max_value = np.max([abs(value) for value in flattened_landmark_list])
    #print(max_value)

    normalized_landmark_list = []

    for value in flattened_landmark_list:
        value /= max_value
        normalized_landmark_list.append(value)
        #print(value)
   

    return normalized_landmark_list
```

```python
def save_data_in_csv(data, label_index):
        print(f"{label_index} pressed")
        print("Saving data")
        with open('data.csv', 'a') as f:
            f.write(str(label_index) + ',')
            for hand_landmarks in data:
                normalized_landmark_list = normalize_landmark_list(hand_landmarks.landmark)
                for value in normalized_landmark_list:
                    f.write(str(value) + ',')
            f.write('\n')
```

```python

mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_drawing = mp.solutions.drawing_utils

cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        continue

    results = hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    cv2.imshow('Hand Tracking', frame)
    
    current_data_label = 18
    if cv2.waitKey(10) & 0xFF == ord('0'):
        print(f"{current_data_label} pressed")
        print("Saving data")
        with open('data.csv', 'a') as f:
            f.write(str(current_data_label) + ',')
            for hand_landmarks in results.multi_hand_landmarks:
                normalized_landmark_list = normalize_landmark_list(hand_landmarks.landmark)
                for value in normalized_landmark_list:
                    f.write(str(value) + ',')
            f.write('\n')
    
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
```
